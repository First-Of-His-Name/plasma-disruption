# Plasma Disruption Prediction using LSTM

#### What are RNNs?
- The idea behind RNNs is to make use of sequential information. In a traditional neural network we assume that all inputs (and outputs) are independent of each other.
- But in some applications, knowing the previous input is very important like in language translations.
- RNNs are called recurrent because they perform the same task for every element of a sequence, with the output being depending on the previous computations.
- RNNs have shown great success in many NLP tasks. Most commonly used type of RNNs are LSTMs, which are much better at capturing long-term dependencies than vanilla RNNs are.

# Syncing Timesteps
<img src="ip_mir_diff_2.png" align="left"/>
<img src="ip_mir_diff_1.png" align="left"/>

<br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br>

- Ip started from <b>-6.4ms</b> and ended at <b>151.4ms</b>, while mirnov started from <b>-6.54ms</b> and ended at <b>122.88ms</b>.
- To sync these readings, we clipped the signals in the range of <b>0</b> to <b>100ms</b>.

```python
start, end = 0.0, 100.0
ip_start, mirnov_start = 0, 0
ip_end, mirnov_end = 0, 0
for i in range(len(ip)):
    if ip[i,0] == start:
        ip_start = i        # storing the start of ip (point where timestamp = 0.0)
    if ip[i,0] == end:
        ip_end = i          # storing the end of ip (point where timestamp = 100.0)
        break
        
for i in range(len(mirnov)):
    if mirnov[i,0] == start:
        mirnov_start = i    # storing the start of ip (point where timestamp = 0.0)
    if int(mirnov[i,0]) == int(end):
        mirnov_end = i      # storing the end of ip (point where timestamp = 100.0)
        break
        
print('ip: {0}-{1} timestep\nmirnov: {2}-{3} timestep'.format(ip_start, ip_end, mirnov_start, mirnov_end))
print('ip value: {0}-{1}'.format(ip[ip_start,0], ip[ip_end,0]))
print('mirnov value: {0}-{1}'.format(mirnov[mirnov_start,0], mirnov[mirnov_end,0]))

# clip all features
ip_clipped = ip[ip_start:ip_end, :]
mirnov_clipped = mirnov[mirnov_start:mirnov_end, :]
bolo_clipped = bolo[ip_start:ip_end, :]
c3_clipped = c3[ip_start:ip_end, :]
o1_clipped = o1[ip_start:ip_end, :]
vloop_clipped = vloop[ip_start:ip_end, :]
ha_clipped = ha[ip_start:ip_end, :]
hxr_clipped = hxr[mirnov_start:mirnov_end, :]
sxr_clipped = sxr[mirnov_start:mirnov_end, :]
delta_clipped = delta[ip_start:ip_end, :]
print('ip length: {0}\nmirnov length: {1}'.format(len(ip_clipped), len(mirnov_clipped)))

# get maximum of mirnov, hxr and sxr
max_mirnov, max_hxr, max_sxr = [], [], []
for i in range(0, len(mirnov_clipped), 25):
    max_mirnov.append(np.max(mirnov_clipped[i:i+25, 1]))
    max_hxr.append(np.max(hxr_clipped[i:i+25, 1]))
    max_sxr.append(np.max(sxr_clipped[i:i+25, 1]))
```

Some probes take data more frequently like hxr, mirnov and sxr. Frequency:
$\frac{12500}{500} = 25$ <br><br>
The sampling rate of Ip was <b>0.2ms</b> while that of Mirnov was <b>0.008ms</b>.

```python
# calculating max, avg, min and decimated value of mirnov16 signal
max_val, max_idx, min_val, min_idx, mean_val = [], [], [], [], []
for i in range(0, mirnov_clipped.shape[0], 25):
    max_val.append(np.amax(mirnov_clipped[i:i+25, 1]))
    max_idx.append(hxr[np.argmax(mirnov_clipped[i:i+25, 1])+i, 0])
    min_val.append(np.amin(mirnov_clipped[i:i+25, 1]))
    min_idx.append(hxr[np.argmin(mirnov_clipped[i:i+25, 1])+i, 0])
    mean_val.append(np.mean(mirnov_clipped[i:i+25, 1]))
    
from scipy.signal import decimate
mirnov_decimated = decimate(mirnov_clipped[:,1],25,1)

plt.figure(figsize=(15,10))

plt.subplot(511)
plt.title("original")
plt.plot(mirnov_clipped[:6250,0], mirnov_clipped[:6250,1])

plt.subplot(512)
plt.title("max")
plt.plot(max_idx[:250], max_val[:250])

plt.subplot(513)
plt.title("min")
plt.plot(min_idx[:250], min_val[:250])

plt.subplot(514)
plt.title("avg")
plt.plot(max_idx[:250], mean_val[:250])

plt.subplot(515)
plt.title("decimated")
plt.plot(mirnov_decimated[:250])

plt.tight_layout()
plt.show()
```

The data of 25 timesteps i.e. 25 consecutive reading were combined to reduce the dimentions of some probes.
This combination can be done is different ways like taking the maximum, minimum, or average of values in 25 timesteps.<br><br>
As seen in the plot above, 'decimated' compresses the original mirnov signal best. But, as the precursor for disruption, the amplitude of mirnov is considered. Therefore, we have taken maximum for a window.

# Correlation between various diagnostics

```python
import seaborn as sns

x = df.drop('label', axis=1).values

# getting average of correlation for all shots
for i in range(119):
    x_reshaped = np.empty((500, 10))
    # reshaping the values
    for j in range(10):
        x_reshaped[:, j] = x[i][j]
    
    dfc = pd.DataFrame(x_reshaped)
    dfc = dfc.rename(columns={0:'bolo', 1:'c3', 2:'delta', 3:'ha', 4:'hxr', 5:'ip', 6:'mirnov', 7:'o1', 8:'sxr', 9:'vloop'})
    sns.set(style="white")
    if i==0:
        corr = dfc.corr()
    else:
        corr = corr.add(dfc.corr())

corr = corr/119

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0.5, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

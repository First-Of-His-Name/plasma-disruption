# Predicting Plasma Disruption by LSTMs

#### What are RNNs?
- The idea behind RNNs is to make use of sequential information. In a traditional neural network we assume that all inputs (and outputs) are independent of each other.
- But in some applications, knowing the previous input is very important like in language translations.
- RNNs are called recurrent because they perform the same task for every element of a sequence, with the output being depended on the previous computations.
- RNNs have shown great success in many NLP tasks. Most commonly used type of RNNs are LSTMs, which are much better at capturing long-term dependencies than vanilla RNNs are.

# Bringing data to same format

Initial lengths:
- bolo: 790
- c3: 790

- <b>hxr: 16179</b>
- ip: 790

- vloop: 790

Some probes take data more frequently like hxr, mirnov and sxr. Frequency:
$\frac{16179}{790} = 20.27$

```python
max_val, max_idx, min_val, min_idx, mean_val = [], [], [], [], []
for i in range(0, hxr.shape[0], 20):
    max_val.append(np.amax(hxr[i:i+20, 1]))
    max_idx.append(hxr[np.argmax(hxr[i:i+20, 1])+i, 0])
    min_val.append(np.amin(hxr[i:i+20, 1]))
    min_idx.append(hxr[np.argmin(hxr[i:i+20, 1])+i, 0])
    mean_val.append(np.mean(hxr[i:i+20, 1]))
    
plt.figure(figsize=(15,8))

plt.subplot(411)
plt.title("original")
plt.scatter(hxr[1000:3600,0], hxr[1000:3600,1])

plt.subplot(412)
plt.title("max")
plt.scatter(max_idx[50:180], max_val[50:180])

plt.subplot(413)
plt.title("min")
plt.scatter(min_idx[50:180], min_val[50:180])

plt.subplot(414)
plt.title("avg")
plt.scatter(max_idx[50:180], mean_val[50:180])

plt.tight_layout()
plt.show()
```

The data of 20 timesteps i.e. 20 consecutive reading were combined to reduce the dimentions of some probes.
This combination can be done is different ways like taking the maximum, minimum, or average of values in 20 timesteps.<br><br>
As seen in the plot above, taking the average mimics the original data efficiently.

```python
max_hxr, max_mirnov, max_sxr = [], [], []
for i in range(0, hxr.shape[0], 20):
    max_hxr.append(np.amax(hxr[i:i+20, 1]))
    max_mirnov.append(np.amax(mirnov[i:i+20, 1]))
    max_sxr.append(np.amax(sxr[i:i+20, 1]))
    
print('Shape after combining timesteps: ')
print("bolo:  ", bolo.shape[0])
print("c3:    ", c3.shape[0])
print("delta: ", delta.shape[0])
print("ha:    ", ha.shape[0])
print("hxr:   ", len(max_hxr))
print("ip:    ", ip.shape[0])
print("mirnov:", len(max_mirnov))
print("o1:    ", o1.shape[0])
print("sxr:   ", len(max_sxr))
print("vloop: ", vloop.shape[0])
```

# Input data
This data was stored in the dataframe, which can be used to train the model.

```python
import gzip, pickle
with gzip.open('dae_feat.pkl', 'rb') as ifp:
    df = pickle.load(ifp)
```

```python
pd.set_option('display.width', 150)
print("How the data looks.")
print(df.head())
df.drop('shot', axis=1, inplace=True)
```

```python
y = df['label'].values
X = df.drop('label', axis=1).values

X_train = np.empty((97,790,10))
y_train = np.empty((97,790,1))

for i in range(X.shape[0]):
    for j in range(X[i].shape[0]):
        X_train[i,:,j] = X[i][j]

for i in range(y.shape[0]):
    y_train[i,:,0] = y[i]

X_test, y_test = X_train[-1].reshape((1,790,10)), y_train[-1].reshape((1,790,1))
X_val, y_val = X_train[-17:-1], y_train[-17:-1]
X_train, y_train = X_train[:-17], y_train[:-17]

print('X_train shape: {0}\ny_train shape: {1}\nX_val shape: {2}\ny_val shape: {3}\nX_test shape: {4}\ny_test shape: {5}'.format(X_train.shape, 
                                 y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape))
```

# Correlation Matrix

```python
import seaborn as sns
df1 = pd.DataFrame(X_train[0])
df1 = df1.rename(columns={0:'bolo', 1:'c3', 2:'delta', 3:'ha', 4:'hxr', 5:'ip', 6:'mirnov', 7:'o1', 8:'sxr', 9:'vloop'})
sns.set(style="white")
corr = df1.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0.5,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# Model
We used a Deep Learning model compposed of single LSTM cell

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, TimeDistributed
```

```python
model = Sequential()
model.add(LSTM(1,input_shape=(740, 10), return_sequences=True))
# model.add(TimeDistributed(Dense(1, activation='sigmoid')))
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

history = model.fit(X_train[:,:740,:], y_train[:,50:,:], batch_size=4, epochs=100, verbose=1, 
                    validation_data=(X_val[:,:740,:], y_val[:,50:,:]))
# scores = model.evaluate(X_test[0,:750,:], y_test[0,40:,:], verbose=0)
# print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Results

```python
pred = model.predict(X_test[:,:740,:], batch_size=1)
```

```python
plt.scatter(np.arange(pred.shape[1]), y_test[:,:740,:], c='b')
plt.scatter(np.arange(pred.shape[1]), pred, c='g')

plt.ylabel('output')
plt.xlabel('time')
plt.legend(['predicted', 'actual'], loc='upper left')
plt.show()
```
